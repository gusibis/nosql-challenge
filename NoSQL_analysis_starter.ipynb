### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.   '$gte'

'''  
The query with $eq or $in operators returns 0 probably because
it does not match exactly LocalAuthorityName 'London'. 
The query with $regex operator returns 33 
most likely because it matches documents where the LocalAuthorityName field contains the word 'London'
'''
# query = {'LocalAuthorityName': {'$eq': 'London'}, 'RatingValue': {'$gte': 4}}
# query = {'LocalAuthorityName': {'$in': 'London'}, 'RatingValue': {'$gte': '4'}}
# t = establishments.find(query)
# pprint(len(list(t)))
# london_loc_auth_count = establishments.count_documents(query)
# london_loc_auth_count
#################################################################################################
##################################################################################################
query2 = {'LocalAuthorityName': {'$regex': "London"},'RatingValue': {'$gte': '4'}}
london_loc_auth = establishments.find(query2)
london_loc_auth_count = establishments.count_documents(query2)
pprint(f"There are {london_loc_auth_count} documents with London as the LocalAuthority & RatingValue >= 4")
##################################################################################################
##################################################################################################
# Another way
london_loc_auth_alt_way = db.establishments.find({'RatingValue': {'$gte': '4'},'LocalAuthorityName': {'$regex': "London"}}) 
# pprint(list(london_loc_auth_alt_way)[0])  # first document
london_loc_auth_alt_way_count = establishments.count_documents({'RatingValue': {'$gte': '4'},'LocalAuthorityName': {'$regex': "London"}})
pprint(f"Again, found {london_loc_auth_alt_way_count} documents with London as the LocalAuthority & RatingValue >= 4")
##################################################################################################
##################################################################################################
# Use count_documents to display the number of documents in the result
# Display the first document in the results using pprint
print()
print("Here is the first document in the results using pprint:")
pprint(london_loc_auth[0])
```

```python
# Convert the result to a Pandas DataFrame
# res_df_london_loc_auth = pd.DataFrame(list(london_loc_auth)) #One method to convert to DataFrame
res_df_london_loc_auth =  pd.json_normalize(list(london_loc_auth)) 
res_df_london_loc_auth
# Display the number of rows in the DataFrame
print(f"Number of rows = {res_df_london_loc_auth.shape[0]} ")
# Display the first 10 rows of the DataFrame
res_df_london_loc_auth.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.

degree_search = 0.01
#####################################################################################################
penang_flavours= establishments.find_one({"BusinessName":"Penang Flavours"})
#####################################################################################################
latitude =  round(float(penang_flavours['geocode']['latitude']), 10)
longitude = round(float(penang_flavours['geocode']['longitude']), 10)
string = "latitude is " + str(latitude) + ", longitude is " + str(longitude) 
print(string)
print()
# pprint(longitude)
######################################################################################################

query = {'RatingValue': {'$eq': '5'},'geocode.longitude': {'$gt':(longitude - degree_search),'$lt':(longitude + degree_search)},
         'geocode.latitude': {'$gt':(latitude - degree_search), '$lt':(latitude + degree_search)}
        }
sort =  [('scores.Hygiene', -1)]

res = (list(establishments.find(query).sort(sort).limit(10)))
# # Print the results
print(f"Results within 0.01 degree of either side of latitude and longitude from the new restaurant 'Penang Flavours' are below ")
pprint(res)
```

```python
# Convert result to Pandas DataFrame
res_df_coordinates_based =  pd.json_normalize(list(res)) 
res_df_coordinates_based
```
